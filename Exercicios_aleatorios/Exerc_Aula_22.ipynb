Exercício 2: Percorra todos os elementos da lista e some 10 aos valores entre 0 e 10,
multiplique por 20 os valores entre 11 e 50, subtraia 30 dos valores entre 51 e
100 e divida por 2 os valores acima de 100.

10, 11, 423, 2034, 134, 11, 3, 23, 34, 758, 9, 10, 14, 2, 412,34, 123,
4132, 4, 12, 3412, 34, 12, 34132, 41, 234, 1, 23,4, 12, 3, 41, 2, 4123

```python
#  Menor que 10, some 10
#  Entre 11 e 50, multiplique por 20
#  Entre 51 e 100, subtraia 30
#  Acima de 100, divida por 2

lista_num = [10, 11, 423, 2034, 134, 11, 3, 23, 34, 758, 9, 10, 14, 2, 412,34, 123, 4132, 4, 12, 3412, 34, 12, 34132, 41, 234, 1, 23,4, 12, 3, 41, 2, 4123]

lista_soma = []
lista_mult = []
lista_sub = []
lista_div = []

for calc in lista_num:
  if calc <= 10:
    lista_soma.append(calc + 10)
  elif calc <= 50:
    lista_mult.append(calc * 20)
  elif calc <= 100:
    lista_sub.append(calc - 30)
  else:
    lista_div.append(calc / 2)
print(lista_soma)
print(lista_mult)
print(lista_sub)
print(lista_mult)
```
