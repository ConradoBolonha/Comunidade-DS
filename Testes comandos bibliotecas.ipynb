```python
carros = {'Marca': 'FIAT', 'Veículo': 'Cronos', 'Cor': 'Prata', 'Ano': 2021}
```

```python
#  ADICIONA O ELEMENTO "MODELO" NA BIBLIOTECA
carros['Modelo'] = 2022
```

```python
# MOSTRA O VALOR ATRIBUIDO AO ELEMENTO "MODELO" 
print(carros['Modelo'])
```

```python
# APAGA UM ELEMENTO DA BIBLIOTECA
del carros['Cor']
```

```python
#  MOSTRA OS VALORES (FIAT, NOME etc)
print(carros.values())
```

```python
#  MOSTRA AS KEYS DA BIBLIOTECA (MARCA, VEÍCULO, COR, etc)
print(carros.keys())
```
